```python
def is_factor(number,index):
    if number % index == 0:
        return True
    return False

def is_prime_number(number):
    if number<2 :
        return False
    index =2
    is_prime=True
    while number >index:
        if is_factor(number,index):
            is_prime = False
            break
        index+=1
    print(is_prime)   


is_prime_number(1)    

        
```
